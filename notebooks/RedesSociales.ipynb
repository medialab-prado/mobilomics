```python
df_insta=pd.read_csv('Instagram_Mobility_20_fixed.csv')
#df_insta.dropna()
df_insta=df_insta.dropna(subset=['coords'])
df_insta.head()
```

```python
len(df_insta)
```

```python
df_geo_insta = pd.DataFrame()
df_geo_insta['id']=df_insta['guid']
df_geo_insta['coords']=df_insta['coords']
df_geo_insta['time']=df_insta['pubdate']
df_geo_insta.head()
```

```python
coords=df_geo_insta['coords']
coords
lat=[]
lon=[]
list=[]
coords[0].split(",")
for c in coords:
    #print c
    c = c.split(",")
    lat.append(c[0])
    lon.append(c[1])
```

```python
df_geo_insta['latitude']=lat
df_geo_insta['longitude']=lon
del df_geo_insta['coords']
```

```python
df_geo_insta.to_csv('instagram20.csv')
```

```python
df_geo_insta.head()
import time
hours=[]
pattern='%Y-%m-%d %H:%M:%S'
for i in df_geo_insta['time']:
    timeData=time.strptime(i, pattern)
    h=timeData.tm_hour
    hours.append(h)
```

```python
#df_geo_insta.to_csv('instagram.csv')
df_geo_insta['hour']=hours
df_geo_insta['hour'].unique()
```

```python
df_geo_insta.to_csv('instagram20.csv')
```
