# GeoJson

```python
import geojson
```

```python
disasterName = '13_Iquique'
disasterRows = df[df['disaster.event']==disasterName]
disasterRows = disasterRows.sample(300)

latitudes  = disasterRows['longitude.anon']
longitudes = disasterRows['latitude']
```

```python
disasterRows
```

```python
centro = (longitudes.mean(),latitudes.mean())
print "centro: {},{}".format(centro[0],centro[1]) 

dest   = (-0.09,51.505) 
offset = (dest[0] - centro[0], dest[1] - centro[1])
```

```python
point_coordinates = zip(longitudes.apply(lambda x: x + offset[0]),latitudes.apply(lambda x: x + offset[1]))
```

```python
multipoint = geojson.MultiPoint(point_coordinates)
```

```python
fh = open(disasterName + '_geo.js','w')
fh.write("var geojsonFeature = ")
geojson.dump(multipoint,fh)
fh.close()
```
